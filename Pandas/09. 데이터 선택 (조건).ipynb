# 데이터 선택 (조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col = '지원번호') # 인덱스 지정
df
```

```python
df['키'] >= 185 # 학생들의 키가 185 이상인지 여부를 True / False
```

```python
filt = (df['키'] >= 185) # 키가 185 이상인 데이터만
df[filt]
```

```python
df[~filt] # filt 를 역으로 적용 (위 데이터를 제외한 나머지 데이터 적용)
```

```python
df[df['키'] >= 185 ]
```

```python
df.loc[df['키'] >= 185, '수학'] # 키가 185 이상인 학생들의 수학 데이터 가져오기
```

```python
df.loc[df['키'] >= 185, ['이름', '수학', '과학']] # 키가 185 이상인 학생들의 이름, 수학, 과학 데이터 가져오기
```

## 다양한 조건


### & 그리고

```python
df.loc[(df['키'] >= 185) & (df['학교'] == '북산고')] # 키가 185 이상이고 북산고 학생인 데이터
```

### | 또는

```python
df.loc[(df['키'] > 200) | (df['키'] < 170)] # 키가 200 보다 크거나 170 보다 작은 학생
```

### str 함수

```python
filt = df['이름'].str.startswith('송') # '송' 씨 성을 가진 사람
df[filt]
```

```python
filt = df['이름'].str.contains('태') # 이름에 '태' 가 들어가는 사람
df[filt]
```

```python
filt = df['이름'].str.contains('태') # 이름에 '태' 가 들어가는 사람
df[~filt] # 이름에 '태' 가 들어가지 않은 사람
```

```python
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs) # SW특기가 Python 이거나 Java 인 사람
df[filt]
```

```python
langs = ['python', 'java']
filt = df['SW특기'].str.lower().isin(langs) # 특기를 소문자로 다 바꿔주어 대소문자 구분을 없앰
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java', na = True) # NaN 데이터에 대해서 True 로 설정
```

```python
filt = df['SW특기'].str.contains('Java', na = False) # NaN 데이터에 대해서 False 로 설정
```
